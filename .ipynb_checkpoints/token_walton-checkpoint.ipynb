```python
current_time = datetime.datetime.utcnow()

t = datetime.timedelta(seconds=0)
cont=0;
data_list=[]
while True:
    #cpu_nodo = [0, 0]
    #mem_nodo = [0, 0]
    cpu_Usage = 0
    mem_Usage = 0
    i = 0
    #info=info_token(access_token)
    expired,e =expire_token(access_token)
    
    
    
    while t.total_seconds()< 10: # delay 20 sec
        time = datetime.datetime.utcnow()
        t=time-current_time
    #cont=cont+1
    if int(e.total_seconds()/60) < 10: #  to generate new_token for x min
        access_token,err=new_token(url,refresh_token)
        print(access_token)
        
    
    t = datetime.timedelta(seconds=0)
    current_time = datetime.datetime.utcnow()
    

   
   
    oscar_cluster="https://inference-walton.cloud.imagine-ai.eu"   
    url_status = oscar_cluster + "/system/status"
   # Make the GET request with basic authentication or token authentication
    
    headers = {
    'Authorization': "Bearer " + access_token,
    'Content-type': 'application/json' 
    }
    timestamp = datetime.datetime.utcnow().strftime("%Y-%m-%d %H:%M:%S")
    response = requests.get(url_status, headers=headers, verify=True)

# Check the status of the response
    if response.status_code == 200:
    # Convert the response to JSON
        try:
            data = response.json()
            print(data)
            if isinstance(data, dict):
                nodos = len(data['detail'])
                data=data['detail']
                
                
             # Iterate over each object in the array
                if nodos >= 1:
                    cpu_nodo=[0]*(nodos)
                    cpu_nodo_percent=[0]*(nodos)
                    mem_nodo=[0]*(nodos)
                    mem_nodo_percent=[0]*(nodos)
                    
                # Iterate over each object in the array, except the front node
                    for obj in data:
                    # Calculate the available CPU and memory (sum of each node)
                    
                        mem_nodo[i] = int(obj['memoryUsage'])
                        cpu_nodo[i] = int(obj['cpuUsage'])
                        cpu_nodo_percent[i]=float(obj['cpuPercentage'])
                        mem_nodo_percent[i]=float(obj['memoryPercentage'])
                        
                        
                        cpu_Usage += cpu_nodo[i]
                        mem_Usage += mem_nodo[i]
                        i += 1
                    
            else:
                print("The response is not a JSON array of objects.")
       
    
        except ValueError as e:
            print("Error converting the response to JSON:", e)
    else:
        print(f"Request error: {response.status_code}")
        print("Error message:")
        print(response.text)

    print(cpu_Usage)
    print(mem_Usage)

 # Crear una fila de datos para agregar a la lista
    resultado = ''
    for k, elemento in enumerate(cpu_nodo):
        resultado += str(cpu_nodo[k]) + ',' +str(cpu_nodo_percent[k])+ ','+ str(mem_nodo[k]) + ','+str(mem_nodo_percent[k])+','
    
    data_row = [timestamp, cpu_Usage, mem_Usage] + cpu_nodo +cpu_nodo_percent+ mem_nodo+mem_nodo_percent
    data_list.append(data_row)
    #print(data_list)
    
    # Crear nombres de columnas dinámicamente
    column_names = ['Timestamp', 'CPU_Usage', 'Mem_Usage']
    for j in range(1, nodos+1):
        column_names.append(f'CPU_Node{j}')
    for j in range(1, nodos+1):
        column_names.append(f'CPU_Node{j}_%')
    for j in range(1, nodos+1):
        column_names.append(f'Mem_Node{j}')
    for j in range(1, nodos+1):
        column_names.append(f'Mem_Node{j}_%')
    #print(column_names)

    # Guardar los valores en un archivo Excel
    df = pd.DataFrame(data_list,columns=column_names)
    df.to_excel("cpu_memoria_18_10.xlsx", index=False)
    #time.sleep(20)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Leer los datos del archivo Excel
file_path = "cpu_memoria_18_10.xlsx"
df = pd.read_excel(file_path)

# Convertir la columna 'Timestamp' a formato de fecha y hora
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Crear una figura para la asignación de CPU y la CPU de cada nodo
plt.figure(figsize=(14, 7))

# Gráfico de CPU Allocation
plt.subplot(2, 1, 1)

plt.plot(df['Timestamp'], df['CPU_Usage'], label='CPU Usage', color='b')
plt.xlabel('Timestamp')
plt.ylabel('CPU Usage')
plt.title('CPU Usage Over Time')
plt.legend()
plt.grid(True)

# Gráfico de CPU de cada nodo
plt.subplot(2, 1, 2)
for i in range(1, nodos+1):
    plt.plot(df['Timestamp'], df[f'CPU_Node{i}'], label=f'CPU Node {i}')
plt.xlabel('Timestamp')
plt.ylabel('CPU Usage per Node')
plt.title('CPU Usage per Node Over Time')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig('cpu_usage.jpg')  # Guardar la imagen en formato JPG
plt.show()

# Crear una figura para la asignación de memoria y la memoria de cada nodo
plt.figure(figsize=(14, 7))

# Gráfico de Memory Allocation
plt.subplot(2, 1, 1)

plt.plot(df['Timestamp'], df['Mem_Usage'] / 1e9, label=' CPU Usage', color='r')
plt.xlabel('Timestamp')
plt.ylabel('Memory Usage')
plt.title('Memory Usage Over Time')
plt.legend()
plt.grid(True)

# Gráfico de memoria de cada nodo
plt.subplot(2, 1, 2)
for i in range(1, nodos+1):
    
    plt.plot(df['Timestamp'], df[f'Mem_Node{i}']/1e9, label=f'Memory Usage {i}')
plt.xlabel('Timestamp')
plt.ylabel('Memory Usage per Node')
plt.title('Memory Usage per Node Over Time')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig('memory_usage.jpg')  # Guardar la imagen en formato JPG
plt.show()
```

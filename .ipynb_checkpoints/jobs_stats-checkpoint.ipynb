```python
import json
from datetime import datetime
import matplotlib.pyplot as plt

with open('job_status_17_10_5000.txt', 'r') as f:
    jobs = f.read()
    print("Contenido del archivo:")
    #print(jobs)
data=json.loads(jobs)
tiempos=list(data.items())

# Listas para almacenar las duraciones
duraciones_creation_to_start = []
duraciones_start_to_finish = []
job_ids = []


for i,data in enumerate(tiempos):
    creation_time = datetime.fromisoformat(data[1]['creation_time'].replace('Z', '+00:00'))
    start_time = datetime.fromisoformat(data[1]['start_time'].replace('Z', '+00:00'))
    finish_time = datetime.fromisoformat(data[1]['finish_time'].replace('Z', '+00:00'))

# Calcular las diferencias
    duracion_creation_to_start = start_time - creation_time
    duracion_start_to_finish = finish_time - start_time

# Obtener las duraciones en segundos
    duracion_creation_to_start_seconds = duracion_creation_to_start.total_seconds()
    duracion_start_to_finish_seconds = duracion_start_to_finish.total_seconds()
    
    duraciones_creation_to_start.append(duracion_creation_to_start_seconds)
    duraciones_start_to_finish.append(duracion_start_to_finish_seconds/3600)
    job_ids.append(i+1)

# Imprimir los resultados
    #print("Duracin entre Creation Time y Start Time en segundos:", duracion_creation_to_start_seconds)
    #print("Duracin entre Start Time y Finish Time en segundos:", duracion_start_to_finish_seconds)
x = range(len(job_ids))  # Eje X: ndices de los trabajos
mean_start = sum(duraciones_creation_to_start) / len(duraciones_creation_to_start)
mean_finish = sum(duraciones_start_to_finish) / len(duraciones_start_to_finish)
print(mean_start)
print(mean_finish)

# Graficar Duracin Creation a Start
plt.figure(figsize=(10, 6))
plt.bar(job_ids, duraciones_creation_to_start, color='b')
plt.axhline(mean_start, color='blue', linestyle='--')
plt.text(x=5, y=mean_start + 0.5, s=f'Mean: {mean_start} sec', color='black')
plt.xlabel('Job IDs')
plt.ylabel('Time between Created and Started (sec)')
plt.title(' Created to Started Time')
plt.xticks(ticks=range(1, len(job_ids) + 1)) # Rotar etiquetas del eje X
plt.tight_layout()
plt.savefig('jobs_start_17_10_5000.jpg')  # Guardar la imagen en formato JPG
plt.show()

# Graficar Duracin Start a Finish
plt.figure(figsize=(10, 6))
plt.bar(job_ids, duraciones_start_to_finish, color='g')
plt.axhline(mean_finish, color='green', linestyle='--')
plt.text(x=5, y=mean_finish + 0.5, s=f'Mean: {round(mean_finish,2)} hour', color='black')
plt.xlabel('Job IDs')
plt.ylabel('Time between Started and Finished (hour)')
plt.title('Started to Finished Time')
plt.xticks(ticks=range(1, len(job_ids) + 1))
#plt.xticks(rotation=45)  # Rotar etiquetas del eje X
plt.tight_layout()
plt.savefig('jobs_processing_17_10_5000.jpg')  # Guardar la imagen en formato JPG
plt.show()
```

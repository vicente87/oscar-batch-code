```python
import jwt
import datetime

# Supongamos que tienes un token JWT (aquí se usa uno de ejemplo)
token=''
# Clave secreta usada para firmar el token (la necesitarás si el token está firmado)
secret_key = ""

try:
    # Decodificar el token (sin verificar la firma)
    decoded_token = jwt.decode(token, options={"verify_signature": False}, algorithms=["HS256"])

    # Extraer el campo de expiración 'exp'
    exp_timestamp = decoded_token.get('exp')

    if exp_timestamp:
        # Convertir el timestamp de expiración a un objeto datetime
        exp_datetime = datetime.datetime.utcfromtimestamp(exp_timestamp)

        # Comparar con la hora actual
        current_time = datetime.datetime.utcnow()

        if current_time >= exp_datetime:
            print("El token ha expirado")
        else:
            print(f"El token es válido. Expira en: {exp_datetime}")
    else:
        print("El token no tiene fecha de expiración")

except jwt.ExpiredSignatureError:
    print("El token ya ha expirado.")
except jwt.InvalidTokenError:
    print("Token inválido.")
```
